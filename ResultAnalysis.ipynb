Ovviamente si puÃ² notare che le correlazioni cambiano da dataset a dataset, proprio perchÃ¨ i dataset sono eterogei tra di loro, per grandezza e contenuti.

Andiamo a fare un'analisi della correlazione, di ogni variabile con l'accuratezza...

```python
import pandas as pd

df_Compas = pd.read_csv("CompasDataset/output/results_correlazione.csv")
df_German = pd.read_csv("GermanDataset/output/results_correlazione.csv")
df_Sms = pd.read_csv("SmsSpamDataset/output/results_correlazione.csv")
```

```python
#stampo i df 
df_German_senza_variabile = df_German["Correlazione con Accuratezza"]
df_Sms_senza_variabile = df_Sms["Correlazione con Accuratezza"]

# Concatena i dataframe insieme, mantenendo solo la colonna "Variabile" all'inizio
df_concatenato = pd.concat([df_Compas, df_German_senza_variabile, df_Sms_senza_variabile], axis=1)

# Mostra il dataframe concatenato
print(df_concatenato)
```

Queste qui sopra sono le correlazioni di ogni variabile con l'accuratezza. Ora ne faccio una media.

```python
# Calcolo della media delle correlazioni per variabile (riga)
df_concatenato['Media_correlazioni'] = df_concatenato.iloc[:, 1:].mean(axis=1)
print(df_concatenato['Media_correlazioni'])
```

In generale, vedendo i risultati, posso concludere che nel mio studio ho notato scarsa correlazione tra accuratezza e le variabili calcolate.
